Ques1:Can we use Bagging for regression problems?

Ques2:What is the difference between multiple model training and single model training?

Ans: Single model training trains one model on the entire datasets for tasks or outputs. While, multiple model training trains multiple model of subsets of datasets for computing the output.

Ques3: Explain the concept of feature randomness in Random Forest.

Ans: In Random Forest, feature randomness means at each split, the algorithm randomly selects a subset of features and chooses the best split among them.This randomness ensures that not all trees rely on the same dominant features, which increases diversity across the forest.

Ques4:What is OOB (Out-of-Bag) Score?

Ans: OOB samples are part of the train data not used in model training but act as validation data for specific individual DecisionTree.The average of all these OOB samples are called OOB score.

Ques5:How can you measure the importance of features in a Random Forest model?

Ans:In a Random Forest, feature importance is measured mainly in two ways:
- by tracking how much each feature reduces impurity across all trees (Mean Decrease in Impurity)
- by shuffling a feature’s values and observing how much the model’s accuracy drops (Permutation Importance).
- The first method is fast and built-in but can be biased, while the second is more reliable though computationally heavier.

Ques6: Explain the working principle of a Bagging Classifier.

Ans: A Bagging Classifier works by training multiple base models on different bootstrap samples of the training data and then combining their predictions through majority voting. Each bootstrap sample is created by randomly selecting data points with replacement, so every model sees a slightly different dataset. This randomness introduces diversity among the models, reducing variance and preventing overfitting. Finally, the ensemble aggregates the predictions—if most models vote for a class, that becomes the final output—making the Bagging Classifier more stable and accurate than a single model.

Ques7: How do you evaluate a Bagging Classifier’s performance?

Ans:  Bagging Classifier’s performance can be evaluated using  metrics like accuracy, precision, recall, F1-score, confusion matrix, and ROC-AUC on a test set. Additionally, Bagging offers the Out-of-Bag (OOB) score, which uses the data left out during bootstrap sampling as a built-in validation set, giving a reliable estimate of performance without needing extra data.

Ques8: How does a Bagging Regressor work?

Ans: A Bagging Regressor works by training multiple regressors on different bootstrap samples of the training data and then averaging their predictions. Each bootstrap sample is created by randomly selecting data points with replacement, so every model sees a slightly different dataset. This diversity reduces variance and prevents overfitting. Finally, the ensemble combines the outputs by averaging them, producing smoother and more robust predictions compared to a single regressor.

Ques9: What is the main advantage of ensemble techniques?

Ans: The main advantage of ensemble techniques is that they combine the strengths of multiple models to produce more accurate, stable, and robust predictions than any single model alone.

Ques10: What is the main challenge of ensemble methods?

Ans: The main challenge of ensemble methods is their complexity and resource cost.

Ques11:  Explain the key idea behind ensemble techniques.

Ans: The key idea behind ensemble techniques is to combine the predictions of multiple models to achieve better performance than any single model could on its own. By aggregating diverse learners—whether through methods like bagging (reducing variance), boosting (reducing bias), or stacking (leveraging different model strengths)—ensembles create a more accurate, stable, and robust predictor.

Ques12: What is a Random Forest Classifier?

Ans:A Random Forest Classifier is an ensemble learning algorithm that builds and combines multiple decision trees to improve classification accuracy and robustness. Each tree is trained on a different bootstrap sample of the data, and at each split, only a random subset of features is considered. This randomness ensures diversity among the trees, reducing overfitting. When making predictions, each tree votes for a class, and the final output is decided by majority voting.

Ques13:  What are the main types of ensemble techniques?

Ans: The main types of ensemble techniques are:
- Bagging: Builds multiple models on different bootstrap samples and combines their outputs (majority vote for classification, averaging for regression) to reduce variance.
- Boosting: Trains models sequentially, with each new model focusing on correcting the errors of the previous ones, reducing bias.
- Stacking: Combines predictions from diverse models using a meta-model that learns how to best blend their outputs.

Ques14: What is ensemble learning in machine learning?

Ans: Ensemble learning in machine learning is the technique of combining multiple models (often called base learners or weak learners) to create a stronger overall model.

Ques15:  When should we avoid using ensemble methods?

Ans: We should avoid using ensemble methods when interpretability and simplicity are critical, since ensembles are harder to explain compared to single models.

Ques16: How does Bagging help in reducing overfitting?

Ans: Bagging helps reduce overfitting by training multiple models on different bootstrap samples of the data and then combining their predictions.

Ques17: Why is Random Forest better than a single Decision Tree?

Ans: A Random Forest is generally better than a single Decision Tree because it reduces overfitting and improves accuracy.

Ques18: What is the role of bootstrap sampling in Bagging?

Ans: The role of bootstrap sampling in Bagging is to create multiple diverse training datasets by randomly sampling the original data with replacement.

Ques19:  What are some real-world applications of ensemble techniques?

- Weather & Science → forecasting and environmental modeling

Ques20: What is the difference between Bagging and Boosting?

Ans: The difference between Bagging and Boosting:

Bagging
- Builds models independently and in parallel on different bootstrap samples of the data.
- Combines predictions by majority voting (classification) or averaging (regression).
- reduce variance and prevent overfitting.

Boosting
- Builds models sequentially, where each new model focuses on correcting the errors of the previous ones.
- Combines predictions by weighted voting or summation, giving more weight to stronger learners.
- reduce bias and improve accuracy.

```python
# 21.  Train a Bagging Classifier using Decision Trees on a sample dataset and print model accuracy
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import BaggingClassifier
from sklearn.metrics import accuracy_score

data = load_iris()
X = data.data
y = data.target

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.30,random_state=1)

base = DecisionTreeClassifier()

model = BaggingClassifier(estimator=base,n_estimators=30,random_state=2)
model.fit(X_train,y_train)

y_pred = model.predict(X_test)
print(accuracy_score(y_test,y_pred))
```

```python
# 22. Train a Bagging Regressor using Decision Trees and evaluate using Mean Squared Error (MSE)
from sklearn.datasets import make_regression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import BaggingRegressor
from sklearn.metrics import mean_squared_error

X,y = make_regression(n_samples=1000,n_features=7,random_state=2)

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=3)

reg = DecisionTreeRegressor()

model = BaggingRegressor(estimator=reg,n_estimators=100,random_state=2)
model.fit(X_train,y_train)

y_pred = model.predict(X_test)
print(mean_squared_error(y_test,y_pred))
```

```python
#  24. Train a Random Forest Regressor and compare its performance with a single Decision Tree
from sklearn.ensemble import RandomForestRegressor

data = load_breast_cancer()
X, y = data.data, data.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=3)

tree_model = DecisionTreeRegressor(random_state=3)
tree_model.fit(X_train, y_train)
tree_preds = tree_model.predict(X_test)
tree_mse = mean_squared_error(y_test, tree_preds)

rf_model = RandomForestRegressor(n_estimators=100, random_state=3)
rf_model.fit(X_train, y_train)
rf_preds = rf_model.predict(X_test)
rf_mse = mean_squared_error(y_test, rf_preds)

print(f"Single Decision Tree MSE: {tree_mse:.4f}")
print(f"Random Forest Regressor MSE: {rf_mse:.4f}")
```

```python
# 31. Train a Random Forest Classifier and tune hyperparameters using GridSearchCV

from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report

iris = load_iris()
X = iris.data
y = iris.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=4)

rf = RandomForestClassifier(random_state=3)

param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [None, 5, 10],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'bootstrap': [True, False]
}

grid_search = GridSearchCV(
    estimator=rf,
    param_grid=param_grid,
    cv=5,
    n_jobs=-1,
    verbose=2
)

grid_search.fit(X_train, y_train)

print("Best Parameters:", grid_search.best_params_)

best_rf = grid_search.best_estimator_
y_pred = best_rf.predict(X_test)
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

```python
# 32. Train a Bagging Regressor with different numbers of base estimators and compare performance

from sklearn.datasets import load_diabetes
from sklearn.ensemble import BaggingRegressor

X, y = load_diabetes(return_X_y=True)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

n_estimators_list = [10, 50, 100, 200]
results = {}

for n in n_estimators_list:
    model = BaggingRegressor(estimator=DecisionTreeRegressor(),n_estimators=n,random_state=42,n_jobs=-1)
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    mse = mean_squared_error(y_test, y_pred)
    results[n] = mse

print("Performance Comparison (MSE):")
for n, mse in results.items():
    print(f"{n} estimators: MSE = {mse:.4f}")
```

```python
# 33.Train a Random Forest Classifier and analyze misclassified samples

from sklearn.metrics import confusion_matrix

iris = load_iris()
X = iris.data
y = iris.target
feature_names = iris.feature_names
target_names = iris.target_names

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)

y_pred = rf.predict(X_test)

print("Classification Report:\n", classification_report(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))

misclassified_indices = [i for i, (true, pred) in enumerate(zip(y_test, y_pred)) if true != pred]

print("\nMisclassified Samples:")
for i in misclassified_indices:
    print(f"Index {i}, True: {target_names[y_test[i]]}, Predicted: {target_names[y_pred[i]]}, Features: {X_test[i]}")
```

```python
# 36.Train a Stacking Classifier using Decision Trees, SVM, and Logistic Regression, and compare accuracy

from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import StackingClassifier


X, y = load_iris(return_X_y=True)

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

base_learners = [
    ('decision_tree', DecisionTreeClassifier(random_state=42)),
    ('svm', SVC(probability=True, random_state=42)),
    ('log_reg', LogisticRegression(max_iter=1000, random_state=42))
]

stacking_clf = StackingClassifier(
    estimators=base_learners,final_estimator=LogisticRegression(max_iter=1000, random_state=42),cv=5)
dt = DecisionTreeClassifier(random_state=42).fit(X_train, y_train)
svm = SVC(probability=True, random_state=42).fit(X_train, y_train)
log_reg = LogisticRegression(max_iter=1000, random_state=42).fit(X_train, y_train)

stacking_clf.fit(X_train, y_train)

acc_dt = accuracy_score(y_test, dt.predict(X_test))
acc_svm = accuracy_score(y_test, svm.predict(X_test))
acc_log = accuracy_score(y_test, log_reg.predict(X_test))
acc_stack = accuracy_score(y_test, stacking_clf.predict(X_test))

print("Decision Tree Accuracy:", acc_dt)
print("SVM Accuracy:", acc_svm)
print("Logistic Regression Accuracy:", acc_log)
print("Stacking Classifier Accuracy:", acc_stack)
```

```python
# 39.Train a Random Forest Classifier and analyze the effect of max_depth on accuracy

X, y = load_iris(return_X_y=True)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=4)

depth_values = [None, 2, 4, 6, 8, 10]
results = {}

for depth in depth_values:
    rf = RandomForestClassifier(
        n_estimators=100,
        max_depth=depth,
        random_state=42
    )
    rf.fit(X_train, y_train)
    y_pred = rf.predict(X_test)
    acc = accuracy_score(y_test, y_pred)
    results[depth] = acc

print("Accuracy with different max_depth values:")
for depth, acc in results.items():
    print(f"max_depth={depth}: Accuracy = {acc:.4f}")

plt.figure(figsize=(6,4))
plt.plot([str(d) for d in depth_values], list(results.values()), marker='o')
plt.xlabel("max_depth")
plt.ylabel("Accuracy")
plt.title("Effect of max_depth on Random Forest Accuracy")
plt.show()
```

```python
# 40.Train a Bagging Regressor using different base estimators (DecisionTree and KNeighbors) and compare
# performance

from sklearn.neighbors import KNeighborsRegressor

X, y = load_diabetes(return_X_y=True)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=4)

bag_dt = BaggingRegressor(estimator=DecisionTreeRegressor(),n_estimators=50,random_state=4)

bag_dt.fit(X_train, y_train)
y_pred_dt = bag_dt.predict(X_test)
mse_dt = mean_squared_error(y_test, y_pred_dt)

bag_knn = BaggingRegressor(estimator=KNeighborsRegressor(),n_estimators=50,random_state=42)

bag_knn.fit(X_train, y_train)
y_pred_knn = bag_knn.predict(X_test)
mse_knn = mean_squared_error(y_test, y_pred_knn)

print("Performance Comparison (MSE):")
print(f"Bagging with Decision Tree: MSE = {mse_dt:.4f}")
print(f"Bagging with KNeighbors:    MSE = {mse_knn:.4f}")
```

```python
# 41.Train a Random Forest Classifier and evaluate its performance using ROC-AUC Score

from sklearn.metrics import roc_auc_score, roc_curve
from sklearn.preprocessing import label_binarize

iris = load_iris()
X, y = iris.data, iris.target
classes = iris.target_names

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=4)

rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)

y_score = rf.predict_proba(X_test)

y_test_bin = label_binarize(y_test, classes=[0, 1, 2])

roc_auc = roc_auc_score(y_test_bin, y_score, average="macro")
print("ROC-AUC Score (macro-average):", roc_auc)

plt.figure(figsize=(6, 5))
for i in range(len(classes)):
    fpr, tpr, _ = roc_curve(y_test_bin[:, i], y_score[:, i])
    plt.plot(fpr, tpr, label=f"{classes[i]} (AUC = {roc_auc_score(y_test_bin[:, i], y_score[:, i]):.2f})")

plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curves - Random Forest Classifier")
plt.legend(loc="lower right")
plt.show()
```

```python
# 43. Train a Random Forest Classifier and plot the Precision-Recall curve.

from sklearn.metrics import precision_recall_curve, average_precision_score
from sklearn.preprocessing import label_binarize

iris = load_iris()
X, y = iris.data, iris.target
classes = iris.target_names

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)

y_score = rf.predict_proba(X_test)

y_test_bin = label_binarize(y_test, classes=[0, 1, 2])

plt.figure(figsize=(7, 5))
for i in range(len(classes)):
    precision, recall, _ = precision_recall_curve(y_test_bin[:, i], y_score[:, i])
    ap = average_precision_score(y_test_bin[:, i], y_score[:, i])
    plt.plot(recall, precision, label=f"{classes[i]} (AP = {ap:.2f})")

plt.xlabel("Recall")
plt.ylabel("Precision")
plt.title("Precision-Recall Curves - Random Forest Classifier")
plt.legend(loc="lower left")
plt.show()
```

```python
# 44.Train a Stacking Classifier with Random Forest and Logistic Regression and compare accuracy

from sklearn.ensemble import RandomForestClassifier, StackingClassifier

X, y = load_iris(return_X_y=True)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=4)

base_learners = [
    ('rf', RandomForestClassifier(n_estimators=100, random_state=42)),
    ('log_reg', LogisticRegression(max_iter=1000, random_state=42))
]

stacking_clf = StackingClassifier(
    estimators=base_learners,
    final_estimator=LogisticRegression(max_iter=1000, random_state=42),
    cv=5
)

rf = RandomForestClassifier(n_estimators=100, random_state=42).fit(X_train, y_train)
log_reg = LogisticRegression(max_iter=1000, random_state=42).fit(X_train, y_train)

stacking_clf.fit(X_train, y_train)

acc_rf = accuracy_score(y_test, rf.predict(X_test))
acc_log = accuracy_score(y_test, log_reg.predict(X_test))
acc_stack = accuracy_score(y_test, stacking_clf.predict(X_test))

print("Random Forest Accuracy:", acc_rf)
print("Logistic Regression Accuracy:", acc_log)
print("Stacking Classifier Accuracy:", acc_stack)
```
