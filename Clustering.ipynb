1. What is unsupervised learning in the context of machine learning?

Ans:Unsupervised learning is a type of machine learning that trains models on unlabeled data, meaning the algorithm is not provided with any pre-defined categories or "correct" answers. Its primary goal is to discover hidden patterns, structures, or groupings within the data on its own. Common applications include clustering (grouping similar data points together), association (identifying relationships between variables), and dimensionality reduction (simplifying complex data while retaining essential information).

2. How does K-Means clustering algorithm work?

Ans:K-Means clustering is an iterative algorithm that partitions a dataset into K distinct groups by minimizing the distance between data points and their respective cluster centers, known as centroids. The process begins by randomly initializing $K$ centroids, then cycles through two main steps: assignment, where each data point is linked to its nearest centroid, and update, where centroids are recalculated as the mean of all points in their group. This cycle repeats until the centroids stabilize, effectively organizing the data into clusters where points within the same group are as similar as possible.

3. Explain the concept of a dendrogram in hierarchical clustering.

Ans: A dendrogram is a tree-like diagram used in hierarchical clustering to visualize the arrangement of clusters and the relationships between data points. It illustrates how individual points are merged into groups (in agglomerative clustering) or how groups are split apart (in divisive clustering).

4. What is the main difference between K-Means and Hierarchical Clustering?

Ans: Here is difference between K-Means and Hierarchical Clustering:

* Number of Clusters ($K$):

K-Means: You must pre-define the number of clusters before running the model.

Hierarchical: No need to pre-define; you decide the number after viewing the results.

* Structure:

K-Means: Creates "flat" groups (data belongs to one specific cluster).

Hierarchical: Creates a tree-like hierarchy (visualized with a dendrogram).

* Efficiency:

K-Means: Faster and works well with large datasets.

Hierarchical: Slower and computationally expensive; best for smaller datasets.

* Flexibility:

K-Means: Typically finds spherical or circular cluster shapes.

Hierarchical: Can find more diverse and complex cluster shapes.

* Consistency:

K-Means: Results can vary depending on where the initial "centers" are placed.

Hierarchical: Results are consistent and reproducible every time.

5.  What are the advantages of DBSCAN over K-Means?

Ans: DBSCAN offers several advantages over K-Means, most notably its ability to identify clusters of arbitrary shapes and sizes, whereas K-Means is restricted to finding spherical clusters. Unlike K-Means, DBSCAN does not require you to pre-specify the number of clusters, as it automatically determines them based on data density. Additionally, DBSCAN is highly robust to outliers and noise, effectively labeling isolated points as noise rather than forcing them into a cluster, which prevents the cluster centers from being skewed. These features make it much more flexible for real-world datasets where the number of groups is unknown and the data contains significant interference.

6.  When would you use Silhouette Score in clustering?

Ans: The **Silhouette Score** is used to evaluate the quality of clustering by measuring how similar an object is to its own cluster compared to other clusters. It is primarily used to determine the **optimal number of clusters** ($K$) by identifying the configuration with the highest score, which indicates well-defined and appropriately separated groups. A score near **+1** suggests that points are perfectly clustered, while a **negative** score warns that data points may be assigned to the wrong group.

7. What are the limitations of Hierarchical Clustering?

Ans: The main limitations of hierarchical clustering are:

* **Computational Cost:** It is very slow and memory-intensive ($O(n^2)$ or $O(n^3)$), making it impractical for large datasets.
* **Irreversibility:** Once two clusters are merged or split, the decision cannot be undone, meaning early errors persist throughout the process.
* **Sensitivity to Noise:** A few outliers or noisy data points can significantly distort the entire tree structure (dendrogram).
* **Subjectivity:** There is no automatic way to determine the "correct" number of clusters; you must manually decide where to "cut" the tree.
* **Difficulty with Shapes:** Depending on the linkage method, it can struggle to identify clusters with complex or irregular shapes.

8. Why is feature scaling important in clustering algorithms like K-Means?

Ans: Feature scaling is crucial in K-Means clustering because the algorithm relies on Euclidean distance to measure similarity.

* If features have very different ranges (e.g., age in years vs. income in thousands), the larger‑scale feature dominates the distance calculation.
* Scaling (like MinMaxScaler or StandardScaler) puts all features on a comparable range, ensuring each contributes fairly.
* Without scaling, clusters may be biased or distorted, leading to poor results.

Scaling prevents one feature from overpowering others, making clusters meaningful and balanced.

9. How does DBSCAN identify noise points?

Ans: DBSCAN marks a point as noise if it doesn’t belong to any cluster.
* Each point needs at least MinPts neighbors within a radius ε (epsilon) to be a core point.
* Points reachable from core points form clusters.
* If a point is not a core and not reachable from any core point, DBSCAN labels it as noise (outlier).

Noise points are those with too few neighbors within ε, so they can’t join any cluster.

10. Define inertia in the context of K-Means.

Ans: Inertia in K-Means is the measure of how well the clusters fit the data. It is calculated as the sum of squared distances between each point and its assigned cluster’s centroid.

11.  What is the elbow method in K-Means clustering?

Ans: The **elbow method** is a technique to decide the optimal number of clusters in K-Means.  
- You run K-Means with different values of \(k\).  
- Plot **inertia (sum of squared distances)** vs. \(k\).  
- The curve usually decreases sharply at first, then flattens.  
- The point where the curve bends like an **elbow** is chosen as the best \(k\).  

It finds the cluster count where adding more clusters stops giving big improvements.

12. Describe the concept of "density" in DBSCAN.

Ans: In DBSCAN, **density** refers to how closely data points are packed together within a region.  
- A region is considered dense if a point has at least **MinPts** neighbors within a radius **ε (epsilon)**.  
- Dense regions form clusters, while sparse regions (low density) are treated as noise.  

Density is the concentration of points within ε distance, used to decide whether a point belongs to a cluster or is noise.

13.  Can hierarchical clustering be used on categorical data?

Ans: Yes, hierarchical clustering can be applied to **categorical data**, but it requires using appropriate **distance measures** (like Hamming distance or Jaccard similarity) instead of Euclidean distance.  
With the right similarity metric, hierarchical clustering can group categorical data meaningfully.

14. What does a negative Silhouette Score indicate?

Ans: A negative Silhouette Score means the point is closer to another cluster than the one it’s assigned to → showing misclassification and poor clustering quality.

15.  Explain the term "linkage criteria" in hierarchical clustering.

Ans: In hierarchical clustering, linkage criteria determine how the distance between clusters is calculated when merging them. Common methods include single linkage (minimum distance between points), complete linkage (maximum distance), and **average linkage** (mean distance). Linkage criteria define the rule for measuring cluster‑to‑cluster distance during the merging process.

16. Why might K-Means clustering perform poorly on data with varying cluster sizes or densities?

Ans: K-Means can perform poorly on data with varying cluster sizes or densities because it assumes clusters are spherical, equally sized, and evenly dense. When clusters differ, points may get wrongly assigned, larger or denser clusters dominate, and the algorithm fails to capture the true structure. K-Means struggles when clusters are unequal in size or density, leading to inaccurate grouping.

17. What are the core parameters in DBSCAN, and how do they influence clustering?

Ans: The core parameters in DBSCAN are:  
- **ε (epsilon):** The radius around a point to search for neighbors.  
- **MinPts:** The minimum number of points required within ε to form a dense region (a core point).  
ε controls neighborhood size, MinPts controls density threshold — together they decide clusters and noise points.

18. How does K-Means++ improve upon standard K-Means initialization?

Ans: K-Means++ improves standard K-Means by choosing initial centroids more intelligently. Instead of picking them randomly, it spreads them out by selecting the first centroid randomly, then choosing subsequent centroids with probability proportional to their distance from already chosen ones.  

19. What is agglomerative clustering?

Ans: Agglomerative clustering is a type of hierarchical clustering that builds clusters from the bottom up. It starts by treating each data point as its own cluster, then repeatedly merges the two closest clusters based on a chosen linkage criteria (like single, complete, or average linkage) until all points are grouped into one big cluster or a stopping condition is met.  

20. What makes Silhouette Score a better metric than just inertia for model evaluation?

Ans: The Silhouette Score is better than inertia because it considers both cluster cohesion (how close points are within the same cluster) and cluster separation (how far clusters are from each other). Inertia only measures compactness, so it can decrease even if clusters overlap.

```python
# 21. Generate synthetic data with 4 centers using make_blobs and apply K-Means clustering. Visualize using a
# scatter plot

import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

X, y = make_blobs(n_samples=500, centers=4, cluster_std=1.0, random_state=42)
kmeans = KMeans(n_clusters=4, random_state=42)
y_kmeans = kmeans.fit_predict(X)

plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, cmap='viridis', s=30)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],c='red', marker='X', s=200, label='Centroids')
plt.title("K-Means Clustering on Synthetic Data")
plt.legend()
plt.show()
```

```python
# 29.  Create synthetic data using make_blobs and evaluate silhouette scores for k = 2 to 5. Display as a bar chart.

import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

X, y = make_blobs(n_samples=500, centers=4, cluster_std=1.0, random_state=42)
scores = {}
for k in range(2, 6):
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(X)
    score = silhouette_score(X, labels)
    scores[k] = score

plt.bar(scores.keys(), scores.values(), color='skyblue')
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Silhouette Score")
plt.title("Silhouette Scores for k = 2 to 5")
plt.show()
```

```python
# 30.  Load the Iris dataset and use hierarchical clustering to group data. Plot a dendrogram with average linkage.

import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from scipy.cluster.hierarchy import dendrogram, linkage

iris = load_iris()
X = iris.data

Z = linkage(X, method='average')
plt.figure(figsize=(10, 6))
dendrogram(Z, truncate_mode=None, p=30, leaf_rotation=90, leaf_font_size=10)
plt.title("Hierarchical Clustering Dendrogram (Average Linkage)")
plt.xlabel("Sample Index")
plt.ylabel("Distance")
plt.show()
```

```python
# 31. Generate synthetic data with overlapping clusters using make_blobs, then apply K-Means and visualize with
# decision boundaries.

import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

X, y = make_blobs(n_samples=500, centers=3, cluster_std=2.0, random_state=42)
kmeans = KMeans(n_clusters=3, random_state=42)
y_kmeans = kmeans.fit_predict(X)

x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.linspace(x_min, x_max, 500),np.linspace(y_min, y_max, 500))

Z = kmeans.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.contourf(xx, yy, Z, alpha=0.3, cmap='viridis')
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, cmap='viridis', s=30)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],c='red', marker='X', s=200, label='Centroids')
plt.title("K-Means Clustering with Decision Boundaries")
plt.legend()
plt.show()
```

```python
# 34.  Load the Breast Cancer dataset and compare inertia values for K = 2 to 6 using K-Means. Show results in a
# line plot.

import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.cluster import KMeans

data = load_breast_cancer()
X = data.data

inertia_values = []
k_values = range(2, 7)
for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X)
    inertia_values.append(kmeans.inertia_)

plt.figure(figsize=(8, 6))
plt.plot(k_values, inertia_values, marker='o', linestyle='-', color='blue')
plt.title("K-Means Inertia Values (Breast Cancer Dataset)")
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Inertia")
plt.grid(True)
plt.show()
```

```python
# 44.  Load the Iris dataset and plot the Silhouette Coefficient for each sample after KMeans clustering.

import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_samples, silhouette_score
import numpy as np

iris = load_iris()
X = iris.data
kmeans = KMeans(n_clusters=3, random_state=42)
labels = kmeans.fit_predict(X)
silhouette_vals = silhouette_samples(X, labels)
avg_score = silhouette_score(X, labels)
print(f"Average Silhouette Score: {avg_score:.3f}")

plt.figure(figsize=(8, 6))
y_lower = 10
for i in range(3):
    ith_cluster_silhouette_values = silhouette_vals[labels == i]
    ith_cluster_silhouette_values.sort()
    size_cluster_i = ith_cluster_silhouette_values.shape[0]
    y_upper = y_lower + size_cluster_i
    color = plt.cm.viridis(float(i) / 3)
    plt.fill_betweenx(np.arange(y_lower, y_upper),
                      0, ith_cluster_silhouette_values,
                      facecolor=color, edgecolor=color, alpha=0.7)
    plt.text(-0.05, y_lower + 0.5 * size_cluster_i, str(i))
    y_lower = y_upper + 10  # gap between clusters

plt.axvline(x=avg_score, color="red", linestyle="--", label="Average Silhouette Score")
plt.title("Silhouette Coefficient for Each Sample (Iris Dataset)")
plt.xlabel("Silhouette Coefficient Values")
plt.ylabel("Cluster Label")
plt.legend()
plt.show()
```

```python
# 46.  Load the Wine dataset, apply KMeans, and visualize the cluster assignments in a seaborn pairplot (first 4
# features).

import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_wine
from sklearn.cluster import KMeans
import pandas as pd

wine = load_wine()
X = wine.data[:, :4]
feature_names = wine.feature_names[:4]
kmeans = KMeans(n_clusters=3, random_state=42)
labels = kmeans.fit_predict(X)
df = pd.DataFrame(X, columns=feature_names)
df['Cluster'] = labels

sns.pairplot(df, vars=feature_names, hue='Cluster', palette='viridis')
plt.suptitle("KMeans Clustering on Wine Dataset (First 4 Features)", y=1.02)
plt.show()
```

```python
# 47.  Generate noisy blobs using make_blobs and use DBSCAN to identify both clusters and noise points. Print the
# count.

import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import DBSCAN
import numpy as np

X, y_true = make_blobs(n_samples=500, centers=4, cluster_std=1.2, random_state=42)
dbscan = DBSCAN(eps=0.8, min_samples=10)
labels = dbscan.fit_predict(X)
n_clusters = len(set(labels)) - (1 if -1 in labels else 0)
n_noise = list(labels).count(-1)
print(f"Number of clusters found: {n_clusters}")
print(f"Number of noise points: {n_noise}")

plt.figure(figsize=(8, 6))
unique_labels = set(labels)
colors = [plt.cm.tab10(each) for each in np.linspace(0, 1, len(unique_labels))]
for k, col in zip(unique_labels, colors):
    if k == -1:
        # Noise points in black
        col = (0, 0, 0, 1)
    class_member_mask = (labels == k)
    plt.scatter(X[class_member_mask, 0], X[class_member_mask, 1], c=[col], s=50, label=f"Cluster {k}" if k != -1 else "Noise")
plt.title("DBSCAN Clustering on Noisy Blobs")
plt.legend()
plt.show()
```

```python
# 48.  Load the Digits dataset, reduce dimensions using t-SNE, then apply Agglomerative Clustering and plot the
# clusters.

import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
from sklearn.manifold import TSNE
from sklearn.cluster import AgglomerativeClustering

digits = load_digits()
X = digits.data
tsne = TSNE(n_components=2, random_state=42, perplexity=30, learning_rate=200)
X_tsne = tsne.fit_transform(X)
agg_cluster = AgglomerativeClustering(n_clusters=10, linkage='ward')
labels = agg_cluster.fit_predict(X_tsne)

plt.figure(figsize=(10, 8))
scatter = plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=labels, cmap='tab10', s=40)
plt.title("Agglomerative Clustering on Digits Dataset (t-SNE 2D)")
plt.xlabel("t-SNE Component 1")
plt.ylabel("t-SNE Component 2")
plt.colorbar(scatter, label="Cluster ID")
plt.show()
```
