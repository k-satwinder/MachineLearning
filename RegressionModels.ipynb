Question 1 : What is Simple Linear Regression (SLR)? Explain its purpose.

Answer: Simple linear regression is a method to study the relationship between two variables by fitting a straight line to data.
- Equation: Y= beta _0 + beta _1X + e
- Purpose:
- Show how one variable (X) affects another (Y)
- Make predictions
- Identify trends
- Quantify the strength of the relationship

Question 2: What are the key assumptions of Simple Linear Regression?

Answer: Here are the key assumptions of linear regression:
- Linearity – The relationship between the independent variable(s) and the dependent variable is linear.
- Independence – Residuals (errors) are independent of each other; no autocorrelation.
- Homoscedasticity – Residuals have constant variance across all levels of the independent variable(s).
- Normality – Residuals are normally distributed.

Question 3: Write the mathematical equation for a simple linear regression model and explain each term.

Answer: The mathematical form of a simple linear regression model is:

Y= beta _0 + beta _1X + e

Explanation of each term:
- Y: Dependent variable (the outcome we want to predict).
- X: Independent variable (the predictor or input).
- beta _0: Intercept — the value of Y when X=0.
- beta _1: Slope — how much Y changes for a one-unit increase in X.
- e : Error term — the difference between the observed value and the predicted value, representing randomness or unexplained variation.

Question 5: What is the method of least squares in linear regression?

Answer: The method of least squares is a mathematical technique used to find the best-fitting line through a set of data points in linear regression. The idea is to minimize the sum of the squared differences (errors) between the observed values and the values predicted by the line.

- Regression Line Equation:
y=beta _0 + beta _1x
- y = dependent variable (response)
- x = independent variable (predictor)
- beta _0 = intercept
- beta _1 = slope
- Residuals:
The difference between the observed value (yi) and the predicted value (y^_i).
- Objective:
Minimize the sum of squared residuals:
S= summition{i=1}^n (observed value - predicted value)^2

Question 7: Name and briefly describe three common evaluation metrics for regression
models.

Answer: Here are three common evaluation metrics for regression models:
- Mean Absolute Error (MAE): Measures the average absolute difference between actual and predicted values.Easy to interpret since it’s in the same unit as the target variable.Treats all errors equally, regardless of direction.
- Mean Squared Error (MSE): Calculates the average of squared differences between actual and predicted values. Penalizes larger errors more heavily due to squaring.Useful when large deviations are particularly undesirable.
- R-squared (Coefficient of Determination): Indicates how well the regression model explains the variability of the target variable.Ranges from 0 to 1, with higher values showing better fit.Provides a measure of the proportion of variance explained by the model.

Question 8: What is the purpose of the R-squared metric in regression analysis?

Answer:
The R-squared (coefficient of determination) metric measures how well a regression model explains the variability of the dependent variable.
- Definition: It represents the proportion of variance in the dependent variable that is explained by the independent variables.
- Range: Values lie between 0 and 1.
- R^2=0: The model explains none of the variability.
- R^2=1: The model perfectly explains all variability.
- Interpretation: A higher R-squared indicates a better fit, meaning the model captures more of the underlying data patterns. However, a very high R-squared does not always mean the model is good—it can sometimes indicate overfitting.
In short, R-squared tells us how much of the variation in the outcome is accounted for by the predictors in the model.

```python
# Question 9: Write Python code to fit a simple linear regression model using scikit-learn
# and print the slope and intercept.
# (Include your Python code and output in the code box below.)

# import libraries
import numpy as np
import pandas as pd

# load dataset
from sklearn.datasets import load_diabetes
data = load_diabetes()

# create dataframe
df = pd.DataFrame(data.data, columns= data.feature_names)
df['target'] = data.target

# seprating data into X and y
X = df[['bmi']]
y = df['target']

# splitting the data
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2, random_state = 1)

# model training
from sklearn.linear_model import LinearRegression

# Create and fit the model
model = LinearRegression()
model.fit(X_train, y_train)

# Print slope (coefficient) and intercept
print("Slope (Coefficient):", model.coef_)
print("Intercept:", model.intercept_)

# predict model
y_pred = model.predict(X_test)
```
