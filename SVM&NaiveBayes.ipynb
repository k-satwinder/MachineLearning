Question1:  What is a Support Vector Machine (SVM)?

Answer: Support Vector Machine (SVM) is a machine learning algorithm that finds the best boundary (hyperplane) to separate data into classes by maximizing the margin between them.
- Support vectors = key data points near the boundary.
- Margin = distance between boundary and support vectors (SVM maximizes this).


Question2: What is the difference between Hard Margin and Soft Margin SVM?

Answer:
Hard Margin SVM
- Assumes the data is perfectly linearly separable (no overlap between classes).
- Finds a hyperplane that separates the classes with the maximum margin and no misclassification.
- Very strict: even one misclassified point makes it fail.
- Works well only when data is clean and noise-free.

Soft Margin SVM
- Allows some misclassification or overlap to handle noisy or non-linearly separable data.
- Introduces a penalty parameter (C) that balances margin size vs. classification errors.
- Large C → fewer misclassifications, tighter margin.
- Small C → wider margin, more tolerance for misclassified points.
- More flexible and practical for real-world datasets.




Question3: What is the mathematical intuition behind SVM?

Answer: The mathematical intuition of an SVM is to find the Maximum Margin Hyperplane—the decision boundary that creates the largest possible "buffer" or "street" between two classes. It defines this boundary using a weight vector (w) and bias (b) through the linear equation w.x+b=0. The model ignores most data points and focuses only on the Support Vectors, which are the closest samples to the boundary that "support" the margin. Mathematically, it solves a constrained optimization problem to minimize ||w||^2 (which maximizes the margin width 2/ ||w|| ) while ensuring every point sits on the correct side of the boundary. If the data isn't linearly separable, it uses a Kernel to project the data into a higher-dimensional space where a flat plane can finally slice through the classes cleanly.

Question4:  What is the role of Lagrange Multipliers in SVM?

Answer:
Role of Lagrange Multipliers
- Convert constraints into optimization: They allow to combine the margin maximization objective with classification constraints into a single function (the Lagrangian).
- Identify support vectors: Only points with non-zero alpha_i become support vectors, they directly influence the hyperplane.
- Dual problem formulation: By using multipliers, SVM is reformulated into a dual optimization problem, which is easier to solve and allows the use of kernels.
- Weight vector representation: The final hyperplane is expressed as:
(w = summition _i(alpha _i.y_i.x_i))
showing that the model depends only on support vectors.


Question5:  What are Support Vectors in SVM?

Answer: Support Vectors in SVM are the data points that lie closest to the decision boundary (hyperplane). They are the critical points that determine the position and orientation of the hyperplane.


Question6: What is a Support Vector Classifier (SVC)?

Answer: Support Vector Classifier(SVC) is the practical implementation of SVM algorithm, it classifies the data into categories by finding the optimal hyperplane that separates them.

Question7: What is a Support Vector Regressor (SVR)?

Answer: Support Vector Regressor (SVR) is the regression predicts the continous values instead of classifying data into categories. It's goal is to fit a function that predicts values while keeping errors within a certain tolerance.

Question8: What is the Kernel Trick in SVM?

Answer: The Kernel Trick in SVM is a way to handle data that can’t be separated by a straight line. It uses special functions (kernels) to map data into higher dimensions, where a simple line can separate it.


Question9:  Compare Linear Kernel, Polynomial Kernel, and RBF Kernel.

Answer:

Linear Kernel
-  K(x,x') = x.x'
- used when data is linearly separable
- simple,fast,works well with high dimensional data
- cannot handle complex, non-linear data

Polynomial Kernel
- K(x,x') = (x.x' + c) ^d
- used when data has curved boundaries
- captures complex relationships
- can be risk of overfitting with high degree

RBF(Radial Based Function) Kernel
- K(x,x') = e^(-|x-x'|^2 / 2sigma^2)
- used when data is highly non-linear and complex
- very flexible, handles almost any shape of boundary
- requires tuning parameter,can be slower on large datasets.


Question10:  What is the effect of the C parameter in SVM?

Answer: The C parameter in SVM controls the trade-off between having a wide margin and allowing classification errors.
Effect of C
- Large C → The model tries to classify all training points correctly.
- Margin becomes smaller (tight boundary).
- Less tolerance for misclassification.
- Risk of overfitting.
- Small C → The model allows some misclassifications.
- Margin becomes wider (looser boundary).
- More tolerance for noisy data.
- Better generalization.


Question11: What is the role of the Gamma parameter in RBF Kernel SVM?

Answer: In RBF Kernel SVM, the Gamma parameter controls how much influence each training example has.
Effect of Gamma
 1. Large gamma  
- Each point has short-range influence.
- Decision boundary becomes very tight and complex.
- Can capture fine details but risks overfitting.
2. Small gamma  
- Each point has long-range influence.
- Decision boundary becomes smoother and simpler.
- Better generalization but may underfit.


Question12: What is the Naïve Bayes classifier, and why is it called "Naïve"?

Answer: The Naïve Bayes classifier is a simple machine learning algorithm based on Bayes’ Theorem that predicts the probability of a data point belonging to a class. It’s called “Naïve” because it assumes all features are independent of each other given the class,- In reality, features often depend on each other.





Question13:  What is Bayes’ Theorem?

Answer: Bayes’ Theorem is a fundamental rule in probability that describes how to update our beliefs when new evidence appears.
Formula
P(A|B)= P(B|A).P(A) / P(B)
Where:
- P(A|B) = probability of event A given evidence B
- P(B|A) = probability of evidence B if A is true
- P(A) = prior probability of A
- P(B) = overall probability of B

- Equal Importance → Every feature contributes equally to the outcome.

Question17:  What are the advantages and disadvantages of Naïve Bayes?

Answer:
 Advantages
- Simple & Fast → Easy to implement and very efficient, even on large datasets.
- Works well with small data → Performs surprisingly well even with limited training samples.
- Effective for text classification → Especially good for spam filtering, sentiment analysis, and document categorization.
- Handles high-dimensional data → Can manage thousands of features (like words in text) without slowing down.
- Probabilistic output → Provides class probabilities, not just labels, which can be useful for decision-making.

 Disadvantages
- Naïve independence assumption → Assumes features are independent, which is often unrealistic.
- Zero-frequency problem → If a feature never appears in training for a class, probability becomes zero (can be fixed with smoothing).
- Limited for continuous data → Gaussian NB works only if data is normally distributed; otherwise performance drops.
- Less flexible → More complex models (like SVM or Random Forest) often outperform it when feature dependencies matter.


Question18:  Why is Naïve Bayes a good choice for text classification?

Answer: Naïve Bayes is a good choice for text classification because text data often involves very high-dimensional features (like thousands of words), and Naïve Bayes handles this efficiently. It works well with sparse data, is fast to train and predict, and performs surprisingly well even with small datasets. The independence assumption, though unrealistic, fits text problems reasonably because word occurrences can often be treated as independent signals. That’s why it’s widely used in spam filtering, sentiment analysis, and document categorization.


Question19: Compare SVM and Naïve Bayes for classification tasks.

 SVM (Support Vector Machine)

- Works well with complex, non-linear data using kernels.
- Usually more accurate but slower to train.

 Naïve Bayes

Question20:  How does Laplace Smoothing help in Naïve Bayes?

Answer: Laplace Smoothing (also called add-one smoothing) helps Naïve Bayes by fixing the zero-frequency problem.

Problem
- If a word/feature never appears in the training data for a class, its probability becomes zero.
- This can wrongly make the whole probability of that class zero during prediction.

Solution
- Laplace smoothing adds a small constant (usually 1) to all feature counts.
- This ensures no probability is ever zero, even for unseen features.

```python
# 23.  Write a Python program to train an SVM Regressor (SVR) on a housing dataset and evaluate it using Mean
# Squared Error (MSE)

from sklearn.datasets import fetch_california_housing
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error

data = fetch_california_housing()
X, y = data.data, data.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=4)

scaler_X = StandardScaler()
scaler_y = StandardScaler()

X_train_scaled = scaler_X.fit_transform(X_train)
X_test_scaled = scaler_X.transform(X_test)

y_train_scaled = scaler_y.fit_transform(y_train.reshape(-1, 1)).ravel()
y_test_scaled = scaler_y.transform(y_test.reshape(-1, 1)).ravel()

svr = SVR(kernel='rbf', C=1.0, epsilon=0.1)
svr.fit(X_train_scaled, y_train_scaled)

y_pred_scaled = svr.predict(X_test_scaled)

mse_scaled = mean_squared_error(y_test_scaled, y_pred_scaled)

y_pred = scaler_y.inverse_transform(y_pred_scaled.reshape(-1, 1))
mse_final = mean_squared_error(y_test, y_pred)

print(f"Mean Squared Error (Scaled): {mse_scaled:.4f}")
print(f"Mean Squared Error (Original Units): {mse_final:.4f}")
```

```python
# 24.  Write a Python program to train an SVM Classifier with a Polynomial Kernel and visualize the decision
# boundary

import numpy as np
import matplotlib.pyplot as plt

iris = datasets.load_iris()
X = iris.data[:, :2]
y = iris.target

X = X[y != 2]
y = y[y != 2]

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

poly_svm = SVC(kernel='poly', degree=3, C=1.0, coef0=1)
poly_svm.fit(X_scaled, y)

h = .02
x_min, x_max = X_scaled[:, 0].min() - 1, X_scaled[:, 0].max() + 1
y_min, y_max = X_scaled[:, 1].min() - 1, X_scaled[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h),np.arange(y_min, y_max, h))

Z = poly_svm.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.figure(figsize=(8, 6))
plt.contourf(xx, yy, Z, cmap=plt.cm.Paired, alpha=0.8)
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=y, edgecolors='k', cmap=plt.cm.Paired)
plt.xlabel('Sepal Length (scaled)')
plt.ylabel('Sepal Width (scaled)')
plt.title('SVM Decision Boundary: Polynomial Kernel (Degree 3)')
plt.show()
```

```python
# 27. Write a Python program to train an SVM Classifier with different C values and compare the decision
# boundaries visually

from sklearn.pipeline import Pipeline

X, y = datasets.make_moons(n_samples=150, noise=0.15, random_state=42)

C_values = [0.01, 1, 100]

fig, axes = plt.subplots(ncols=3, figsize=(18, 5), sharey=True)

def plot_predictions(clf, axes):
    x0s = np.linspace(-1.5, 2.5, 100)
    x1s = np.linspace(-1.0, 1.5, 100)
    x0, x1 = np.meshgrid(x0s, x1s)
    X_new = np.c_[x0.ravel(), x1.ravel()]
    y_pred = clf.predict(X_new).reshape(x0.shape)
    y_decision = clf.decision_function(X_new).reshape(x0.shape)
    axes.contourf(x0, x1, y_pred, cmap=plt.cm.brg, alpha=0.2)
    axes.contourf(x0, x1, y_decision, cmap=plt.cm.brg, alpha=0.1)

for idx, C_val in enumerate(C_values):
    plt.sca(axes[idx])

    svm_clf = Pipeline([("scaler", StandardScaler()),("linear_svc", SVC(kernel="rbf", gamma=5, C=C_val))])

    svm_clf.fit(X, y)

    plot_predictions(svm_clf, axes[idx])
    plt.scatter(X[:, 0][y==0], X[:, 1][y==0], color='blue', alpha=0.5)
    plt.scatter(X[:, 0][y==1], X[:, 1][y==1], color='green', alpha=0.5)
    plt.title(f"Decision Boundary (C = {C_val})")
    plt.axis([-1.5, 2.5, -1.0, 1.5])

plt.tight_layout()
plt.show()
```

```python
# 34. Write a Python program to train an SVM Classifier and a Naïve Bayes Classifier on the same dataset and
# compare their accuracy

import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score

iris = load_iris()
X, y = iris.data, iris.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

svm_model = SVC(kernel='linear')
svm_model.fit(X_train_scaled, y_train)
svm_pred = svm_model.predict(X_test_scaled)

nb_model = GaussianNB()
nb_model.fit(X_train_scaled, y_train)
nb_pred = nb_model.predict(X_test_scaled)

svm_acc = accuracy_score(y_test, svm_pred)
nb_acc = accuracy_score(y_test, nb_pred)

print(f"SVM Accuracy:         {svm_acc * 100:.2f}%")
print(f"Naïve Bayes Accuracy: {nb_acc * 100:.2f}%")
```

```python
# 35.Write a Python program to perform feature selection before training a Naïve Bayes classifier and compare
# results

from sklearn.feature_selection import SelectKBest, f_classif

wine = load_wine()
X, y = wine.data, wine.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=4)

nb_full = GaussianNB()
nb_full.fit(X_train, y_train)
pred_full = nb_full.predict(X_test)
acc_full = accuracy_score(y_test, pred_full)

selector = SelectKBest(score_func=f_classif, k=2)
X_train_selected = selector.fit_transform(X_train, y_train)
X_test_selected = selector.transform(X_test)

nb_selected = GaussianNB()
nb_selected.fit(X_train_selected, y_train)
pred_selected = nb_selected.predict(X_test_selected)
acc_selected = accuracy_score(y_test, pred_selected)

mask = selector.get_support()
best_features = np.array(wine.feature_names)[mask]

print(f"Accuracy with ALL features (13): {acc_full * 100:.2f}%")
print(f"Accuracy with BEST features (2):  {acc_selected * 100:.2f}%")
print(f"The 2 features chosen: {best_features}")
```

```python
# 37.  Write a Python program to train an SVM Classifier using Linear, Polynomial, and RBF kernels on the Breast
# Cancer dataset and compare their accuracy


from sklearn.datasets import load_breast_cancer

data = load_breast_cancer()
X, y = data.data, data.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=5)

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

kernels = ['linear', 'poly', 'rbf']
results = {}

for k in kernels:
    model = SVC(kernel=k, gamma='auto')
    model.fit(X_train, y_train)

    predictions = model.predict(X_test)
    acc = accuracy_score(y_test, predictions)
    results[k] = acc

print("--- SVM Kernel Accuracy Comparison ---")
for k, acc in results.items():
    print(f"{k.capitalize()} Kernel: {acc * 100:.2f}%")
```

```python
# 38.Write a Python program to train an SVM Classifier using Stratified K-Fold Cross-Validation and compute the
# average accuracy.

from sklearn.model_selection import StratifiedKFold, cross_val_score

data = load_breast_cancer()
X, y = data.data, data.target

svm_pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('svm', SVC(kernel='rbf', C=1.0))
])

skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

scores = cross_val_score(svm_pipeline, X, y, cv=skf, scoring='accuracy')

print("--- Stratified K-Fold Results ---")
for i, score in enumerate(scores):
    print(f"Fold {i+1} Accuracy: {score:.4f}")

print(f"\nAverage Accuracy: {np.mean(scores):.4f}")
print(f"Standard Deviation: {np.std(scores):.4f}")
```

```python
# 40. Write a Python program to perform Recursive Feature Elimination (RFE) before training an SVM Classifier and
# compare accuracy

from sklearn.feature_selection import RFE

X, y = load_iris(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

svm = SVC(kernel="linear", random_state=42)
svm.fit(X_train, y_train)
y_pred_baseline = svm.predict(X_test)
baseline_acc = accuracy_score(y_test, y_pred_baseline)

selector = RFE(estimator=svm, n_features_to_select=2, step=1)
selector.fit(X_train, y_train)

X_train_rfe = selector.transform(X_train)
X_test_rfe = selector.transform(X_test)

svm_rfe = SVC(kernel="linear", random_state=42)
svm_rfe.fit(X_train_rfe, y_train)
y_pred_rfe = svm_rfe.predict(X_test_rfe)
rfe_acc = accuracy_score(y_test, y_pred_rfe)

print("Baseline SVM Accuracy (all features):", round(baseline_acc, 4))
print("SVM Accuracy after RFE (2 features):", round(rfe_acc, 4))
print("Selected Features (True=kept):", selector.support_)
```
